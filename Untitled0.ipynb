```python
# --- Step 1: Import Libraries ---
import pandas as pd
import matplotlib.pyplot as plt

# Make plots look nicer
plt.style.use("seaborn-v0_8")
plt.rcParams["figure.figsize"] = (8,5)

# --- Step 2: Load CSV File ---
# Upload sales.csv in Colab first (Files > Upload)
df = pd.read_csv("sales.csv")

# --- Step 3: Explore Data ---
print("First 5 rows of data:")
print(df.head(), "\n")

print("Basic Info:")
print(df.info(), "\n")

print("Summary Statistics:")
print(df.describe(), "\n")

# --- Step 4: Add Total Sales Column ---
df["TotalSales"] = df["Quantity"] * df["UnitPrice"]
print("Data with TotalSales column added:")
print(df.head(), "\n")

# --- Step 5: Analysis ---

# 1. Total Revenue
total_revenue = df["TotalSales"].sum()
print("üí∞ Total Revenue:", total_revenue, "\n")

# 2. Sales by Product
sales_by_product = df.groupby("Product")["TotalSales"].sum().sort_values(ascending=False)
print("üì¶ Sales by Product:\n", sales_by_product, "\n")

# 3. Sales by Region
sales_by_region = df.groupby("Region")["TotalSales"].sum().sort_values(ascending=False)
print("üåç Sales by Region:\n", sales_by_region, "\n")

# 4. Monthly Sales Trend
df["Date"] = pd.to_datetime(df["Date"])  # ensure Date is in datetime format
sales_by_month = df.groupby(df["Date"].dt.to_period("M"))["TotalSales"].sum()
print("üìà Monthly Sales Trend:\n", sales_by_month, "\n")

# --- Step 6: Visualization ---

# Product-wise sales (Bar chart)
sales_by_product.plot(kind="bar", color="skyblue", title="Sales by Product")
plt.ylabel("Revenue")
plt.show()

# Region-wise sales (Pie chart)
sales_by_region.plot(kind="pie", autopct="%1.1f%%", title="Sales by Region")
plt.ylabel("")
plt.show()

# Monthly sales trend (Line chart)
sales_by_month.plot(kind="line", marker="o", color="green", title="Monthly Sales Trend")
plt.ylabel("Revenue")
plt.show()
```
